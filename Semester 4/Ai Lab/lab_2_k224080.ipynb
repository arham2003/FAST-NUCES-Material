```python
import random


def initialize_room(rows, cols):
    room = [['D' if random.random() < 0.3 else 'C' for _ in range(cols)] for _ in range(rows)]
    return room

def display_room(room, vacuum_location):
    for i in range(len(room)):
        for j in range(len(room[0])):
            if (i, j) == vacuum_location:
                print("*", end=" ")  # Representing vacuum location with *
            else:
                print(room[i][j], end=" ")
        print()

def is_all_clean(room):
    return all(all(cell == 'C' for cell in row) for row in room)

def clean_room(rows, cols):
    room = initialize_room(rows, cols)
    vacuum_location = (random.randint(0, rows - 1), random.randint(0, cols - 1))

    print("Initial Room:")
    display_room(room, vacuum_location)

    movements = 0
    while not is_all_clean(room) and movements < 10:
        x, y = vacuum_location
        neighbors = [(x-1, y), (x+1, y), (x, y-1), (x, y+1)]
        dirty_neighbors = [neighbor for neighbor in neighbors if 0 <= neighbor[0] < rows and 0 <= neighbor[1] < cols and room[neighbor[0]][neighbor[1]] == 'D']

        if dirty_neighbors:
            next_location = random.choice(dirty_neighbors)
            room[next_location[0]][next_location[1]] = 'C'
            vacuum_location = next_location
            print(f"\nMovement {movements + 1}: Cleaned cell {next_location}")
        else:
            print("\nNo dirty cells in neighbors. Stopping execution.")
            break

        display_room(room, vacuum_location)
        movements += 1

    print("\nCleaning completed!")

clean_room(3, 3) #room size
```

Write a Python NLTK program to tokenize words, sentence wise.

```python
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize

paragraph = """Lorem ipsum dolor sit amet, consectetur adipiscing elit.
          Vestibulum ullamcorper lacus at lectus hendrerit, non rhoncus turpis vulputate.
          Quisque et nunc ac justo fermentum tristique. In hac habitasse platea dictumst.
          Proin sit amet sapien et lacus finibus bibendum. Aliquam vel elit vel justo posuere pellentesque vel at erat. Maecenas fermentum elit quis odio fringilla lacinia. Nullam eleifend consequat lectus, nec finibus odio dignissim a. Integer auctor odio at mi accumsan facilisis.
          Sed gravida sapien eu justo varius, a bibendum ligula malesuada."""

sentences_list = sent_tokenize(paragraph) #converting paragraph into tokens of sentences
words_list = [word_tokenize(word) for word in sentences_list] #converting sentences into tokens of words

print("List of sentences:", sentences_list)
print("\nList of words for every sentence:")
for i, words in enumerate(words_list, 1):
    print(f"Sentence {i}: {words}")
```
