## Question 1:

```python
def probability_of_fair_given_heads(p_fair_heads, p_unfair_heads, p_select_fair):
    p_heads = p_fair_heads * p_select_fair + p_unfair_heads * (1 - p_select_fair)
    p_fair_given_heads = (p_fair_heads * p_select_fair) / p_heads

    return p_fair_given_heads


p_fair_heads = 0.5
p_unfair_heads = 0.8
p_select_fair = 0.5  # Since both coins are randomly selected, the probability of selecting the fair coin is 0.5

result = probability_of_fair_given_heads(p_fair_heads, p_unfair_heads, p_select_fair)
print("Probability of selecting the fair coin given heads P(Fair coin | heads):", result)
```

## Question 3

```python
from pgmpy.models import BayesianNetwork
from pgmpy.inference import VariableElimination
from pgmpy.factors.discrete import TabularCPD

model = BayesianNetwork([('Attendance', 'ExamPerformance'),
                       ('StudyTime', 'ExamPerformance'),
                       ('PreviousScores', 'ExamPerformance')])

cpd_attendance = TabularCPD(variable='Attendance', variable_card=2, values=[[0.85], [0.15]])  # 0 for poor, 1 for good
cpd_study_time = TabularCPD(variable='StudyTime', variable_card=2, values=[[0.7], [0.3]])  # 0 for less, 1 for more
cpd_previous_scores = TabularCPD(variable='PreviousScores', variable_card=2, values=[[0.6], [0.4]])  # 0 for low, 1 for high

cpd_exam_performance = TabularCPD(variable='ExamPerformance', variable_card=2,
                                  values=[[0.1, 0.4, 0.7, 0.9, 0.2, 0.5, 0.8, 0.95],
                                          [0.9, 0.6, 0.3, 0.1, 0.8, 0.5, 0.2, 0.05]],
                                  evidence=['Attendance', 'StudyTime', 'PreviousScores'], evidence_card=[2, 2, 2])

model.add_cpds(cpd_attendance, cpd_study_time, cpd_previous_scores, cpd_exam_performance)

inference = VariableElimination(model)

passing_probab = inference.query(variables=['ExamPerformance'], evidence={'Attendance': 1, 'StudyTime': 1})
print(passing_probab)
```

## Question 4

```python
outcomes = [(i, j) for i in range(1, 7) for j in range(1, 7)]

sum_of_seven = sum(1 for outcome in outcomes if sum(outcome) == 7)
total_outcomes = len(outcomes)

probability = sum_of_seven / total_outcomes

print(f"The probability of the sum being 7 is {probability:.2f} or {probability*100:.0f}%.")
```

## Question 5:

```python
from pgmpy.models import BayesianNetwork
from pgmpy.inference import VariableElimination
from pgmpy.factors.discrete import TabularCPD

model = BayesianNetwork([('StudyTime', 'Performance'), ('Tutoring', 'Performance')])
cpd_study_time = TabularCPD(variable='StudyTime', variable_card=3, values=[[0.3], [0.4], [0.3]], state_names={'StudyTime': ['low', 'medium', 'high']})
#                                                                Probabilities: low, medium, high

cpd_tutoring = TabularCPD(variable='Tutoring', variable_card=2,values=[[0.7], [0.3]], state_names={'Tutoring': ['no', 'yes']})

cpd_performance = TabularCPD(variable='Performance', variable_card=3,
                             values=[[0.1, 0.05, 0.02, 0.2, 0.1, 0.05],  # for poor
                                     [0.3, 0.4, 0.3, 0.4, 0.5, 0.35],   # for average
                                     [0.6, 0.55, 0.68, 0.4, 0.4, 0.6]],  # for good
                   evidence=['StudyTime', 'Tutoring'], evidence_card=[3, 2], state_names={'Performance': ['poor', 'average', 'good'], 'StudyTime': ['low', 'medium', 'high'],'Tutoring': ['no', 'yes']})

model.add_cpds(cpd_study_time, cpd_tutoring, cpd_performance)

inference = VariableElimination(model)

# Question 1
prob_poor_performance = inference.query(variables=['Performance'], evidence={'StudyTime': 'medium', 'Tutoring': 'yes'})
print("Probability of poor performance:", prob_poor_performance.values[0])
print()

# Question 2
prob_high_study_good_perf = inference.query(variables=['StudyTime'], evidence={'Performance': 'good', 'Tutoring': 'yes'})
print("Probability of high study time given good performance:", prob_high_study_good_perf.values[2])
print()

# Question 3
prob_dist_average_perf = inference.query(variables=['StudyTime', 'Tutoring'], evidence={'Performance': 'average'})
print("Probability distribution of study time and tutoring given average performance:\n", prob_dist_average_perf)
```
