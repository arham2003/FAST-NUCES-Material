**Q1)** Take input from user a dictionary. User should input key as well as value. Now write a code that should have these functionalities :

Asks user to Press 1 if user wants to check "value" associated with any key and If user presses 1 then take input a key and return its associated value
Asks user to Press 2 if user wants to check "key" associated with any value and If user presses 2 then take input a value and return its associated key

```python
#Task 1  22k-4080
global userDict
userDict = {}
n=3

for i in range(n):
  data = input("Enter Key and Value seperated by : ")
  temp = data.split(":")
  userDict[temp[0]]= temp[1]

def checkKey(userKey):
  for key, value in userDict.items():
    if  userKey == key:
      print("The value is : ", value)

def checkValue(userValue):
  for key, value in userDict.items():
    if  userValue == key:
      print("The key is : ", key)

print(userDict)

optPress = int(input("Press 1 to input a key to find its associated value\nPress 2 to input a value to find its associated key:\n"))

if(optPress == 1):
  keyInput = input("Enter a key: \n")
  checkKey(keyInput)

if(optPress == 2):
  valueInput = input("Enter a Value: \n")
  checkValue(valueInput)
```

Q6) Take two integer numbers from user as input. Divide these two numbers. If one number is being divided by zero (another number) then handle ZeroDivisionError and if entered value is wrong (for example, a string) then handle ValueError.

```python
#Task 6

try:
  num1 = int(input("Enter the first integer: "))
  num2 = int(input("Enter the second integer: "))

  result = num1 / num2

  print(f"The result of {num1} / {num2} is {result}")

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

except ValueError:
    print("Error: Please enter valid integer values.")
```
