# TASK 2 (KNN Classifier written for 3 distance metrics)

```python
import numpy as np
import statistics
```

```python
class KNN_Classifier():


    def __init__(self, distance_metric):

      self.distance_metric= distance_metric

    def compute_distance(self, training_data_point, test_data_point):

      if (self.distance_metric.lower() == 'euclidean'):

        dist = 0
        for i in range(len(training_data_point) - 1):
          dist += (training_data_point[i] - test_data_point[i])**2


        euclidean_dist = np.sqrt(dist)
        return euclidean_dist

      elif (self.distance_metric.lower() == 'manhattan'):

        dist = 0

        for i in range(len(training_data_point) - 1):
          dist += abs(training_data_point[i] - test_data_point[i])

        manhattan_dist = dist
        return manhattan_dist

      elif (self.distance_metric.lower() == 'chi-squared'):

        dist = 0

        for i in range(len(training_data_point) - 1):
            if (training_data_point[i] + test_data_point[i]) != 0:
                dist += ((training_data_point[i] - test_data_point[i]) ** 2) / (training_data_point[i] + test_data_point[i])
        chi_squared_dist = dist
        return chi_squared_dist


    def nearest_neighbor(self, X_train, test_data, k): #getting nearest neighbours

        distance_list = []

        for training_data in X_train:

          distance = self.compute_distance(training_data, test_data)
          distance_list.append((training_data, distance))

        distance_list.sort(key= lambda x:x[1]) # i am sorting distance list based on distance value


        neighbors_list = []

        for j in range(k):
            neighbors_list.append(distance_list[j][0])

        return neighbors_list

    def predict(self, X_train, test_data, k):

      neighbors = self.nearest_neighbor(X_train, test_data, k)

      for data in neighbors:
        label = []
        label.append(data[-1]) #target column is last so I have put -1

      predicted_class = statistics.mode(label)

      return predicted_class

```

```python
from sklearn import datasets
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

iris = datasets.load_iris()
X = iris.data
y = iris.target

data = np.column_stack((X, y))
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.2, random_state=42)

classifier = KNN_Classifier(distance_metric='Chi-squared')

y_pred = []
for test_data in X_test:
    y_pred.append(classifier.predict(X_train, test_data, k=5))

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy*100:.2f}%')

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n")
print(conf_matrix)
```
