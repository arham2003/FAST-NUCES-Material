```python
extraction = df20.loc['550': '900']
extraction
```

```python
gend_to_cough = extraction.loc[:, "Gender":"Coughing of Blood"]
gend_to_cough
```

```python
import numpy as np
from scipy import stats

mean = np.mean(gend_to_cough)
print("Mean:",mean.round(2))
median = np.median(gend_to_cough)
print ("Median: ", median)
mode = stats.mode(gend_to_cough)
print("Mode: ", mode)
std_dev = np.std(gend_to_cough, axis = 0)
print("Standard Deviation: ", std_dev)
minimum = np.min(gend_to_cough)
print("minimum: ", minimum)
maximum = np.max(gend_to_cough)
print("Maximum: ", maximum)
```

## part A:

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(nrows=6, ncols=4, figsize=(20, 20))
axes = axes.flatten()

for i, col in enumerate(df20.columns[4:-1]):  #skipping 'index', 'Patient Id', 'Age', and 'Level' columns
    sns.barplot(x='Level', y=col, data=df20, ax=axes[i])
    axes[i].set_title(f'{col} vs Level')
    axes[i].set_ylabel('Count')
    axes[i].set_xlabel('Level')

plt.tight_layout()
plt.show()
```

## PART 3B:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

numeric_df = df20.select_dtypes(include=[int, float]) # Selecting numeric columns for correlation

correlation_matrix = numeric_df.corr()

plt.figure(figsize=(16, 12))

sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)

plt.title('Correlation Heatmap of Lung Cancer Dataset', fontsize=18)
plt.show()
```

## PART 3C:

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 5))

# distribution of 'Age'
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
sns.histplot(df20['Age'], bins=20, kde=True, color='mediumaquamarine', shrink=0.8)
plt.title('Distribution of Age')
plt.xlabel('Age Range')
plt.ylabel('Frequency')

# distribution of 'Frequent Cold'
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
sns.countplot(x='Frequent Cold', data=df20, palette='flare', legend=False)
plt.title('Distribution of Frequent Cold')
plt.xlabel('Frequent Cold Level')
plt.ylabel('Count')

plt.show()
```
