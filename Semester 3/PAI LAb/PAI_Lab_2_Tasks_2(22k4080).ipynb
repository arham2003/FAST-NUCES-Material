Q1) Muneeb has got 40 Marks in Physics, 78 in Chemistry and 82 in Maths. Take these marks as input from user and store them in dictionary with "key as subject name" and "value as marks". By accessing data stored in dictionary, print average of his marks and also print the name of subject in which he got highest marks

```python
#Q1 22k-4080

marksA = float(input("Enter marks for physics: "))
marksB = float(input("Enter marks for chemistry: "))
marksC = float(input("Enter marks for Maths: "))

subjMarks = {
    'Physics' : marksA,
    'Chemistry' : marksB,
    'Maths' : marksC
}

highest = 0
highestsubj = ''
sum = 0
count = 0
for key, value in subjMarks.items():
  sum=sum + value
  count += 1
  if value > highest:
    highest = value
    highestsubj = key

average = sum/count
print(f'\nAverage Marks of Muneeb: {round(average,2)}')


print(f'Highest marks are in {highestsubj}: {highest}')
```

Q3) Write a program to make function employee() meeting following requirements:

Employee name and monthly salary will be passed to this function. This function will cut 2 percent tax from salary and display salary after tax with name of employee. If the salary is missing in the function call then assign default value 10,000 to salary.

```python
#Q3 22k-4080



def employee(Name, Salary):
  if(Salary is None):
    employee(Name,10000)

  tax = float(Salary*0.02)

  Salary -= tax

  print(f'Salary after Tax: {Salary}')


def main():
  empName = input('Enter Employee Name: ')
  empSalary = int(input('Enter Employee Salary: '))
  employee(empName, empSalary)

if __name__ == '__main__':
  main()
```

Q4) Write a function that will take a number 'n' input by user and will return its factorial.

```python
#Q4 22k-4080

def factorial(n):
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result



n = int(input("Enter a non-negative integer: "))
if n < 0:
    print("Factorial is not defined for negative numbers.")
else:
    result = factorial(n)
    print(f"The factorial of {n} is {result}")

```

Q5) Take Name and age as input from user for three persons. Write a function that will take these two arguments i-e Name and age of person. Function will create a dictionary with name as key and age as value.

```python
#Q5 22k-4080

name = []
age = []
for i in range(3):
  print(f"Person {i+1}: \n")
  names = input("Enter Name: ")
  age1 = int(input("Enter Age: "))
  name.append(names)
  age.append(age1)

def createDict(name, age):
  myDict = {}

  for i in range(len(name)):
      myDict[name[i]] = age[i]

  print(myDict)

createDict(name, age)
```
