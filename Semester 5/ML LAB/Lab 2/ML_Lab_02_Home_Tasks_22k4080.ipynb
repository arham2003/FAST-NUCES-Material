```python
def correlation(dataset, threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
      for j in range(i):
        if abs(corr_matrix.iloc[i,j]) > threshold:
          colname = corr_matrix.columns[i]
          col_corr.add(colname)
    return col_corr

corr_features = correlation(df4, 0.65)
print("Dropped Features: ",len(set(corr_features)))
df4.drop(corr_features, axis=1)
```

## Part H: (Feature Selection Using MAD)

```python
# Feature Selection using Mean Absolute Difference

df5 = pd.read_csv("train.csv")

import pandas as pd
import numpy as np

def feature_selection_mad(dataset, threshold):
    features_to_drop = []
    for column in dataset.columns:
        mean_value = np.mean(dataset[column])
        mad_value = np.mean(np.abs(dataset[column] - mean_value)) #Mean Absolute Difference (MAD) for the column

        # If MAD is less than the threshold, add the column to the list of features to drop
        if mad_value < threshold:
            features_to_drop.append(column)

    return features_to_drop

mad_features_to_drop = feature_selection_mad(df5, 0.65)

print(f"Number of features to drop based on MAD: {len(mad_features_to_drop)}")
df5 = df5.drop(columns=mad_features_to_drop)
print("Shape of DataFrame after feature selection using MAD:", df5.shape)
```

### PART E (Handling Imbalanced Data using Combination of Over and Under Sampling)

```python
missing_values = survey_df.isnull().sum()
print("Missing Values per column: \n", missing_values)
```

```python
import pandas as pd
from imblearn.combine import SMOTEENN
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_selector as selector

# Assuming survey_df is your DataFrame and 'Depression' is the target column

# Convert the target variable to categorical if it's continuous
threshold = survey_df['Depression'].median()
survey_df['Depression'] = (survey_df['Depression'] > threshold).astype(int)

# Separate features and target variable
X = survey_df.drop('Depression', axis=1)
y = survey_df['Depression']

# Identify categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns

# Create a column transformer with preprocessing for both numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),  # Handle missing values
            ('scaler', StandardScaler())  # Standardize numerical features
        ]), selector(dtype_exclude='object')),
        ('cat', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),  # Handle missing values
            ('onehot', OneHotEncoder(handle_unknown='ignore'))  # One-hot encode categorical features
        ]), categorical_cols)
    ],
    remainder='passthrough'  # Include remaining columns without transformations
)

# Apply the column transformer to preprocess the data
X_preprocessed = preprocessor.fit_transform(X)

# Initialize SMOTEENN
smote_enn = SMOTEENN(random_state=42)

# Apply SMOTEENN to the preprocessed training data
X_resampled, y_resampled = smote_enn.fit_resample(X_preprocessed, y)

# Display the shape of the resampled data
print("Shape of resampled features:", X_resampled.shape)
print("Shape of resampled target:", y_resampled.shape)

# Check the class distribution after resampling
print("Class distribution after resampling:")
print(pd.Series(y_resampled).value_counts())
```

## Recursive Feature ELimination (Part F):

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X = survey_df.drop('Depression', axis=1) # 'Depression' is the target column
y = survey_df['Depression']

categorical_cols = X.select_dtypes(include=['object']).columns

preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),  # Handle missing values
            ('scaler', StandardScaler())  # Standardize numerical features
        ]), selector(dtype_exclude='object')),
        ('cat', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),  # Handle missing values
            ('onehot', OneHotEncoder(handle_unknown='ignore'))  # One-hot encode categorical features
        ]), categorical_cols)
    ],
    remainder='passthrough'  # Include remaining columns without transformations
)

X_preprocessed = preprocessor.fit_transform(X)

model = LogisticRegression(max_iter=1000, random_state=42)

rfe = RFE(estimator=model, n_features_to_select=10)

rfe.fit(X_preprocessed, y)
ranking = rfe.ranking_ # Getting the ranking of features

selected_features = [feature for feature, rank in zip(X.columns, ranking) if rank == 1]
print("Selected features:", selected_features)
```

## Part F Explanation:

- ColumnTransformer: Handles numerical and categorical data separately.
- Recursive Feature Elimination uses a specified model (LogisticRegression) to determine feature importance.
- "Ranking" Provides a ranking of features, where a rank of 1 indicates the most important features.
