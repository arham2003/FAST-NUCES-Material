```python
#AI TASK 1
import random
import string

TARGET = "Artificial Intelligence Lab"
POPULATION_SIZE = 70

def generate_random_string(length):
    return ''.join(random.choice(string.ascii_letters + string.digits + string.punctuation + ' ') for _ in range(length))

def calculate_fitness_score(individual):
    return sum(1 for i, j in zip(individual, TARGET) if i != j)

def generate_initial_population(population_size):
    return [generate_random_string(len(TARGET)) for _ in range(population_size)]

def selection(population):
    return sorted(population, key=calculate_fitness_score)

def crossover(parent1, parent2):
    crossover_point = random.randint(1, len(TARGET) - 1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

def mutation(individual, mutation_rate):
    mutated_individual = ''
    for gene in individual:
        if random.random() < mutation_rate:
            mutated_individual += random.choice(string.ascii_letters + string.digits + string.punctuation + ' ')
        else:
            mutated_individual += gene
    return mutated_individual

def gene_algorithm():
    population = generate_initial_population(POPULATION_SIZE)
    generation = 1
    while True:
        population = selection(population)
        if population[0] == TARGET:
            return population[0], generation
        new_population = []
        for _ in range(POPULATION_SIZE // 2):
            parent1, parent2 = random.choices(population[:10], k=2)  # Select top 10 individuals as parents
            child1, child2 = crossover(parent1, parent2)
            child1 = mutation(child1, mutation_rate=0.05)
            child2 = mutation(child2, mutation_rate=0.05)
            new_population.extend([child1, child2])
        population = new_population
        print(f"Generation {generation}: Best individual: {population[0]}")
        generation += 1

result, generation = gene_algorithm()
print("Target string '{}' generated in generation {}".format(result, generation))
```

```python
import random

my_cities = ['A', 'B', 'C', 'D']
my_distances = {
    ('A', 'B'): 2,
    ('A', 'C'): 3,
    ('A', 'D'): 1,
    ('B', 'C'): 4,
    ('B', 'D'): 2,
    ('C', 'D'): 3,
    ('B', 'A'): 2,
    ('C', 'A'): 3,
    ('D', 'A'): 1,
    ('C', 'B'): 4,
    ('D', 'B'): 2,
    ('D', 'C'): 3
}

# Initialize population
def my_initialize_population(population_size):
    population = []
    for _ in range(population_size):
        route = random.sample(my_cities, len(my_cities))
        population.append(route)
    return population


def my_route_distance(route):
    total_distance = 0
    for i in range(len(route) - 1):
        total_distance += my_distances[(route[i], route[i+1])]
    total_distance += my_distances[(route[-1], route[0])]  # Complete the loop
    return total_distance

# Fitness function
def my_fitness(route):
    return 1 / my_route_distance(route)

# Selection
def my_selection(population):
    return random.choices(population, weights=[my_fitness(route) for route in population], k=2)

# Crossover
def my_crossover(parent1, parent2):
    crossover_point = random.randint(1, len(my_cities) - 1)
    child = parent1[:crossover_point]
    child.extend([city for city in parent2 if city not in child])
    return child

# Mutation
def my_mutation(route):
    mutation_point = random.randint(1, len(route) - 2)
    route[mutation_point], route[mutation_point + 1] = route[mutation_point + 1], route[mutation_point]
    return route


def my_genetic_algorithm(population_size, generations):
    population = my_initialize_population(population_size)
    for _ in range(generations):
        new_population = []
        for _ in range(population_size):
            parent1, parent2 = my_selection(population)
            child = my_crossover(parent1, parent2)
            if random.random() < 0.1:  # Mutation probability
                child = my_mutation(child)
            new_population.append(child)
        population = new_population
    return min(population, key=my_route_distance)


if __name__ == "__main__":
    my_population_size = 6
    my_generations = 100
    best_route = my_genetic_algorithm(my_population_size, my_generations)
    print("Best route:", best_route)
    print("Distance:", my_route_distance(best_route))
```

```python
#TASK 3
class Node:
    def __init__(self, value, parent=None, heuristic=None):
        self.value = value
        self.parent = parent
        self.heuristic = heuristic
        self.children = []

    def add_child(self, child):
        self.children.append(child)

# Define the tree structure
root = Node("S")
root.add_child(Node("A", root, 3))
root.add_child(Node("B", root, 2))
root.children[0].add_child(Node("C", root.children[0], 4))
root.children[0].add_child(Node("D", root.children[0], 1))
root.children[1].add_child(Node("E", root.children[1], 3))
root.children[1].add_child(Node("F", root.children[1], 1))
root.children[1].children[1].add_child(Node("I", root.children[1].children[1], 2))
root.children[1].children[1].add_child(Node("G", root.children[1].children[1], 3))
root.children[1].children[0].add_child(Node("H", root.children[1].children[0], 5))

def greedy_search(root, destination):
    current_node = root
    visited_nodes = [current_node.value]
    while current_node.value != destination:
        if not current_node.children:
            # Backtrack to the parent node and explore other branches
            while current_node.parent is not None:
                parent = current_node.parent
                # Remove the current node from the list of children
                parent.children.remove(current_node)
                # Check if there are other children nodes to explore
                if parent.children:
                    # Select the child node with the lowest heuristic value
                    current_node = min(parent.children, key=lambda node: node.heuristic)
                    visited_nodes.append(current_node.value)
                    break
                else:
                    # Move up to the parent node
                    current_node = parent
            else:
                # If we reach the root node and still cannot find the destination, it means the destination is not reachable
                return None, visited_nodes
        else:
            # Select the child node with the lowest heuristic value
            current_node = min(current_node.children, key=lambda node: node.heuristic)
            visited_nodes.append(current_node.value)
    return current_node, visited_nodes

# Perform the greedy search from the starting node S to the destination node H
destination_node, visited_nodes = greedy_search(root, 'H')

if destination_node:
    print("Optimal move:", visited_nodes)
    print("Destination node found:", destination_node.value)
else:
    print("Destination node H not reachable!")
```

```python
import heapq

class Node:
    def __init__(self, name):
        self.name = name
        self.g = float('inf')  # Cost from start to node
        self.h = 0  # Heuristic value
        self.f = float('inf')  # Total cost
        self.parent = None  # Parent node in the path
        self.neighbors = {}  # Adjacent nodes and their costs

    def __lt__(self, other):
        return self.f < other.f

def a_star(start, goal, graph):
    open_list = []
    heapq.heappush(open_list, (start.f, start))

    while open_list:
        current = heapq.heappop(open_list)[1]
        if current.name == goal.name:
            return reconstruct_path(current)

        for neighbor_name, cost in current.neighbors.items():
            neighbor = graph[neighbor_name]
            temp_g = current.g + cost

            if temp_g < neighbor.g:
                neighbor.parent = current
                neighbor.g = temp_g
                neighbor.f = temp_g + neighbor.h
                heapq.heappush(open_list, (neighbor.f, neighbor))

    return None

def reconstruct_path(node):
    path = []
    while node:
        path.append(node.name)
        node = node.parent
    return path[::-1]  # Return reversed path

#heuristic values for each node
heuristics = {'S': 5, 'A': 3, 'B': 4, 'C': 2, 'D': 6, 'G': 0}

# Create graph nodes
graph = {name: Node(name) for name in heuristics.keys()}

# Set heuristic values
for name, h_value in heuristics.items():
    graph[name].h = h_value

# Define graph connections and their costs
graph['S'].neighbors = {'A': 2, 'G': 10}
graph['A'].neighbors = {'S': 2, 'B': 2, 'C': 1}
graph['B'].neighbors = {'D': 5}
graph['C'].neighbors = {'D': 3, 'G': 4}
graph['D'].neighbors = {'G': 1}
graph['G'].neighbors = {}

# Set start and goal nodes
start = graph['S']
goal = graph['D']


start.g = 0
start.f = start.g + start.h

path = a_star(start, goal, graph)

if path:
    print("Path found:", " -> ".join(path))
else:
    print("No path found")
```

```python
#TASK 5

import heapq

class Node:
    def __init__(self, position, parent=None):
        self.position = position
        self.parent = parent
        self.g = 0  # Cost from start node
        self.h = 0  # Heuristic value
        self.f = 0  # Total cost

    def __lt__(self, other):
        return self.f < other.f

def heuristic(node, end):
    # Use Manhattan distance as heuristic
    return abs(node.position[0] - end.position[0]) + abs(node.position[1] - end.position[1])

def greedy_best_first_search(start, end, maze):
    open_list = []
    heapq.heappush(open_list, (start.h, start))

    while open_list:
        current = heapq.heappop(open_list)[1]
        if current.position == end.position:
            return reconstruct_path(current)

        neighbors = get_neighbors(current, maze)
        for neighbor in neighbors:
            neighbor.h = heuristic(neighbor, end)
            heapq.heappush(open_list, (neighbor.h, neighbor))

    return None

def a_star_search(start, end, maze):
    open_list = []
    heapq.heappush(open_list, (start.f, start))

    while open_list:
        current = heapq.heappop(open_list)[1]
        if current.position == end.position:
            return reconstruct_path(current)

        neighbors = get_neighbors(current, maze)
        for neighbor in neighbors:
            temp_g = current.g + maze[neighbor.position[0]][neighbor.position[1]]
            if temp_g < neighbor.g:
                neighbor.parent = current
                neighbor.g = temp_g
                neighbor.h = heuristic(neighbor, end)
                neighbor.f = temp_g + neighbor.h
                heapq.heappush(open_list, (neighbor.f, neighbor))

    return None

def get_neighbors(node, maze):
    neighbors = []
    directions = [(0, -1), (-1, 0), (0, 1), (1, 0)]  # Up, Left, Down, Right
    for direction in directions:
        neighbor_pos = (node.position[0] + direction[0], node.position[1] + direction[1])
        if 0 <= neighbor_pos[0] < len(maze) and 0 <= neighbor_pos[1] < len(maze[0]) and maze[neighbor_pos[0]][neighbor_pos[1]] != 0:
            neighbors.append(Node(neighbor_pos, node))
    return neighbors

def reconstruct_path(node):
    path = []
    while node:
        path.append(node.position)
        node = node.parent
    return path[::-1]  # Return reversed path

def main():
    maze = [[10, 9, 8, 7, 6, 5],
            [11, 0, 0, 0, 0, 4],
            [12, 0, 0, 0, 0, 3],
            [13, 0, 0, 0, 0, 2],
            [14, 0, 0, 0, 0, 1],
            [15, 0, 0, 0, 0, 0]]

    start = Node((5, 0))  # Bottom left corner (A)
    end = Node((0, 5))  # Top right corner (B)
    start.h = heuristic(start, end)
    start.f = start.g + start.h

    # Call the search algorithms
    path_greedy = greedy_best_first_search(start, end, maze)
    path_a_star = a_star_search(start, end, maze)

    # Output the paths
    print("Path found by Greedy Best First Search:", path_greedy)
    print("Path found by A* Best First Search:", path_a_star)

if __name__ == "__main__":
    main()
```

```python
import heapq

class PuzzleState:
    def __init__(self, board, parent=None, action=None, path_cost=0):
        self.board = board
        self.parent = parent
        self.action = action
        self.path_cost = path_cost
        self.goal = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
        self.h = self.manhattan_distance()
        self.f = self.path_cost + self.h

    def __lt__(self, other):
        return self.f < other.f

    def manhattan_distance(self):
        distance = 0
        for i in range(3):
            for j in range(3):
                if self.board[i][j] != 0:
                    x, y = divmod(self.board[i][j] - 1, 3)
                    distance += abs(x - i) + abs(y - j)
        return distance

    def is_goal(self):
        return self.board == self.goal

    def get_neighbors(self):
        neighbors = []
        zero_row, zero_col = self.find(0)
        directions = [('up', -1, 0), ('down', 1, 0), ('left', 0, -1), ('right', 0, 1)]

        for action, dx, dy in directions:
            new_row, new_col = zero_row + dx, zero_col + dy
            if 0 <= new_row < 3 and 0 <= new_col < 3:
                new_board = [row[:] for row in self.board]
                new_board[zero_row][zero_col], new_board[new_row][new_col] = new_board[new_row][new_col], new_board[zero_row][zero_col]
                neighbors.append(PuzzleState(new_board, self, action, self.path_cost + 1))

        return neighbors

    def find(self, value):
        for i, row in enumerate(self.board):
            for j, val in enumerate(row):
                if val == value:
                    return i, j

def a_star(initial_board):
    start_state = PuzzleState(initial_board)
    frontier = []
    heapq.heappush(frontier, start_state)
    explored = set()

    while frontier:
        current_state = heapq.heappop(frontier)
        if current_state.is_goal():
            return current_state

        explored.add(tuple(map(tuple, current_state.board)))
        for neighbor in current_state.get_neighbors():
            if tuple(map(tuple, neighbor.board)) not in explored:
                heapq.heappush(frontier, neighbor)

    return None

def print_solution(solution):
    path = []
    while solution:
        path.append(solution.action)
        solution = solution.parent
    return path[::-1]

# initial board
initial_board = [
    [1, 2, 3],
    [0, 4, 6],
    [7, 5, 8]
]

solution = a_star(initial_board)
if solution:
    print("Solution found!")
    print("Path to solution:", print_solution(solution))
else:
    print("No solution found.")
```
