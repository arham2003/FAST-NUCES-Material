```python
#TASK 1

def calc_totalDist(path, distances):
    distance = 0
    num_cities = len(path)
    for i in range(num_cities):
        distance += distances[path[i % num_cities]][path[(i + 1) % num_cities]]
    return distance

def find_shortest_path(distances):
    num_cities = len(distances)

    shortest_path = None
    shortest_distance = float('inf')

    # Generate all permutations of city indices
    for path in generate_permut(num_cities):
        distance = calc_totalDist(path, distances)
        if distance < shortest_distance:
            shortest_distance = distance
            shortest_path = path

    return shortest_path, shortest_distance

#generate permutations of city indices
def generate_permut(n):
    if n == 0:
        return [[]]
    permutations = []
    for perm in generate_permut(n - 1):
        for i in range(n):
            permutations.append(perm[:i] + [n - 1] + perm[i:])
    return permutations


city_distances = [
    [0, 10, 15, 20],
    [10, 0, 21, 27],
    [15, 35, 0, 30],
    [18, 25, 30, 0]
]


shortest_path, shortest_distance = find_shortest_path(city_distances)
print("Shortest Path:", shortest_path)
print("Distance of that shortest path:", shortest_distance)
```

```python
#TASK 2 DFS on graph

def dfsGraph(i, visited, matrix):
    print(i, end=" ")  # Print the current node
    visited[i] = 1  # Mark the current node as visited

    for j in range(len(matrix)):
        if matrix[i][j] != 0 and not visited[j]:
            dfsGraph(j, visited, matrix)  # Recursive call to visit adjacent node


visited = [0, 0, 0, 0]
matrix = [[0, 10, 15, 20],
          [10, 0, 35, 25],
          [15, 35, 0, 30],
          [20, 25, 30, 0]]

# Starting DFS from node 1
dfsGraph(1, visited, matrix)
```

```python
# TASK 2 DFS on tree

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

    def add_child(self, child):
        self.children.append(child)


def dfs_tree(node):
    print(node.data, end=" ")

    for child in node.children:
        dfs_tree(child)


root = TreeNode('A')
b = TreeNode('B')
c = TreeNode('C')
d = TreeNode('D')
e = TreeNode('E')

root.add_child(b)
root.add_child(c)
b.add_child(d)
b.add_child(e)

print("DFS traversal starting from root 'A':")
dfs_tree(root)
```
