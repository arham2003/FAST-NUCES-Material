# Task 2 (same Dataset)

```python
df_attr1 = df[['restecg', 'oldpeak']]
df_label_attr = df['target']
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(df_attr1, df_label_attr, test_size=0.2, random_state=0)

decision_tree = DecisionTreeClassifier(random_state=0)
knn = KNeighborsClassifier()
random_forest = RandomForestClassifier(random_state=0)
xgboost = XGBClassifier(random_state=0, eval_metric='mlogloss')

voting_classify = VotingClassifier(
    estimators=[
        ('dt', decision_tree),
        ('knn', knn),
        ('rf', random_forest),
        ('xgb', xgboost)],
    voting='hard' )  ## hard voting for majority class prediction

voting_classify_soft = VotingClassifier(
    estimators=[
        ('dt', decision_tree),
        ('knn', knn),
        ('rf', random_forest),
        ('xgb', xgboost)],
    voting='soft' )

voting_classify.fit(X_train, y_train) #hard
voting_classify_soft.fit(X_train, y_train) #soft

train_pred = voting_classify.predict(X_train)
test_pred = voting_classify.predict(X_test)

train_pred2 = voting_classify_soft.predict(X_train)
test_pred2 = voting_classify_soft.predict(X_test)

train_acc = accuracy_score(y_train, train_pred)
test_acc = accuracy_score(y_test, test_pred)

train_acc2 = accuracy_score(y_train, train_pred2)
test_acc2 = accuracy_score(y_test, test_pred2)

print("Hard Voting: \n")
print(f"Voting Classifier Training Acc:\n {train_acc*100:.2f}%")
print(f"Voting Classifier Testing Acc:\n {test_acc*100:.2f}%")
print("\n")
print("Soft Voting: \n")
print(f"Voting Classifier Training Acc:\n {train_acc2*100:.2f}%")
print(f"Voting Classifier Testing Acc:\n {test_acc2*100:.2f}%")
```

## Best Weights using weighted classifier

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'weights': [
        [1, 1, 1, 1],      # Equal weights
        [1, 2, 1, 1],      # More weight for KNN
        [2, 1, 1, 1],      # More weight for Decision Tree
        [1, 1, 2, 1],      # More weight for Random Forest
        [1, 1, 1, 2],      # More weight for XGBoost
        [2, 1, 2, 1],      # Combined higher weights for Decision Tree and Random Forest
        [1, 2, 1, 2],      # Combined higher weights for KNN and XGBoost
        [3, 1, 1, 1],      # Much higher weight for Decision Tree
    ]
}

grid_search = GridSearchCV(voting_classify, param_grid, scoring='accuracy', cv=5)
grid_search.fit(X_train, y_train)

best_weights = grid_search.best_params_['weights']
best_score = grid_search.best_score_

voting_classify.set_params(weights=best_weights)
voting_classify.fit(X_train, y_train)
test_acc = accuracy_score(y_test, voting_classify.predict(X_test))

print("Best Weights: ",best_weights*100)
print("Best Cross-Validated Accuracy: ",best_score*100)
print("Test Accuracy with Best Weights: ",test_acc*100)
```

## Bias variance Tradeoff graph

```python
train_accuracies = []
test_accuracies = []

weights = [
    [1, 1, 1, 1],
    [2, 1, 1, 1],
    [1, 2, 1, 1],
    [1, 1, 2, 1],
    [1, 1, 1, 2],
    [3, 1, 1, 1],
    [1, 1, 3, 1],
]
for w in weights:
    voting_clf = VotingClassifier(
        estimators=[
            ('dt', decision_tree),
            ('knn', knn),
            ('rf', random_forest),
            ('xgb', xgboost)
        ],
        voting='soft',
        weights=w
    )

    voting_clf.fit(X_train, y_train)

    train_acc = accuracy_score(y_train, voting_clf.predict(X_train))
    test_acc = accuracy_score(y_test, voting_clf.predict(X_test))

    train_accuracies.append(train_acc)
    test_accuracies.append(test_acc)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(weights) + 1), train_accuracies, label='Training Accuracy', marker='o')
plt.plot(range(1, len(weights) + 1), test_accuracies, label='Testing Accuracy', marker='o')

plt.xlabel('Model Complexity (Varying Weights)', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Bias-Variance Tradeoff - Voting Classifier', fontsize=15)
plt.xticks(ticks=range(1, len(weights) + 1), labels=[str(w) for w in weights], rotation=45)
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
