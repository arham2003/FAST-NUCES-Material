```python
import math
def dist_apart(cor1,cor2):
    x1,y1=cor1
    x2,y2=cor2
    dist = math.sqrt((y2-y1)*2+(x2-x1)*2)
    return round(dist,2)

#assuming the locations of the points other than R
cameraPntR =(4,0)
pntB=(6,0)
pntC=(5,6)
pntN=(10,15)
pntA=(6,10)

print("Distance of A",dist_apart(cameraPntR,pntA))
print("Distance of B",dist_apart(cameraPntR,pntB))
print("Distance of C",dist_apart(cameraPntR,pntC))
print("Distance of N",dist_apart(cameraPntR,pntN))
```

```python
import random

class UniqueAutoCar:
    def __init__(self):
        self.front = 8
        self.left = 2
        self.right = 2
        self.back = 0.05

    def brake(self):
        print("Applying Brakes!")

    def move_left(self):
        print("Moving left")

    def move_right(self):
        print("Moving right")

    def move(self, front, left, right, back):
        self.forward = front
        self.lef = left
        self.rit = right
        self.bac = back

        if self.forward <= self.front:
            print("OH NO! Object ahead")
            self.brake()

        if self.lef <= self.left:
            print("Object at the left")
            self.move_left()

        if self.rit <= self.right:
            print("Object at the right")
            self.move_right()

        if self.bac <= self.back:
            print("There is Object behind")
            self.brake()

print("Car1")
car1 = UniqueAutoCar()
car1.move(7, 1, 1, 0.5)

print("\nCar2")
car2 = UniqueAutoCar()
car2.move(6, 1, 1.5, 0.03)
```

```python
#TASK 4

n = 7 # number of rows
m = 8 # number of columns
room = [
    ["D", "D", "D", "D", "D", "D","D","B"],
    ["B", "B", "C", "D", "B", "B","B","B"],
    ["D", "D", "C", "D", "B", "D","D","D"],
    ["D", "D", "D", "D", "C", "C","C","D"],
    ["D", "D", "D", "D", "D", "D","D","D"],
    ["D", "D", "D", "D", "D", "D","D","D"],
    ["D", "D", "D", "B", "B", "D","D","D"]
]

class Vacuum:

    def __init__(self):
        import random
        self.x = random.randint(0, n-1) # row index
        self.y = random.randint(0, m-1) # column index
        self.path = [(self.x, self.y)]

    # Define a method to check if a cell is valid and dirty
    def is_valid_and_dirty(self, i, j):
        return 0 <= i < n and 0 <= j < m and room[i][j] == "D"

    def move(self, direction):
        if direction == "up":
            self.x -= 1
        elif direction == "down":
            self.x += 1
        elif direction == "left":
            self.y -= 1
        elif direction == "right":
            self.y += 1
        self.path.append((self.x, self.y))
        room[self.x][self.y] = "C"


    def can_move(self):

        return (self.is_valid_and_dirty(self.x-1, self.y) or
                self.is_valid_and_dirty(self.x+1, self.y) or
                self.is_valid_and_dirty(self.x, self.y-1) or
                self.is_valid_and_dirty(self.x, self.y+1))

    def run(self):

        while self.can_move():
            import random
            possible_directions = []
            if self.is_valid_and_dirty(self.x-1, self.y):
                possible_directions.append("up")
            if self.is_valid_and_dirty(self.x+1, self.y):
                possible_directions.append("down")
            if self.is_valid_and_dirty(self.x, self.y-1):
                possible_directions.append("left")
            if self.is_valid_and_dirty(self.x, self.y+1):
                possible_directions.append("right")
            direction = random.choice(possible_directions)
            self.move(direction)

vacuum = Vacuum()
vacuum.run()
print("The path of the vacuum is:")
for cell in vacuum.path:
    print(cell)
print("The final state of the room is:")
for row in room:
    print(row)
```

```python
#TASK 5
import random
class TicTacToeGame:
    def __init__(self):
        self.board = [' '] * 9
        self.player_symbol = 'X'
        self.computer_symbol = 'O'

    def print_board(self):
        print("-------------")
        for i in range(3):
            print("|", self.board[i*3], "|", self.board[i*3+1], "|", self.board[i*3+2], "|")
            print("-------------")

    def check_winner(self, symbol):

        for i in range(3):
            if self.board[i*3] == self.board[i*3+1] == self.board[i*3+2] == symbol:
                return True
            if self.board[i] == self.board[i+3] == self.board[i+6] == symbol:
                return True
        if self.board[0] == self.board[4] == self.board[8] == symbol:
            return True
        if self.board[2] == self.board[4] == self.board[6] == symbol:
            return True
        return False

    def user_move(self):
        while True:
            try:
                move = int(input("Enter your move (1-9): ")) - 1
                if 0 <= move <= 8 and self.board[move] == ' ':
                    self.board[move] = self.player_symbol
                    break
                else:
                    print("Invalid move. Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")

    def computer_move(self):

        for i in range(9):
            if self.board[i] == ' ':
                self.board[i] = self.computer_symbol
                if self.check_winner(self.computer_symbol):
                    return
                else:
                    self.board[i] = ' '


        while True:
            move = random.randint(0, 8)
            if self.board[move] == ' ':
                self.board[move] = self.computer_symbol
                break

    def play_game(self):
        print("Welcome to Tic Tac Toe!")
        self.print_board()
        while ' ' in self.board:
            self.user_move()
            self.print_board()
            if self.check_winner(self.player_symbol):
                print("You won!")
                return
            if ' ' not in self.board:
                print("It's a draw!")
                return
            self.computer_move()
            print("Computer's move:")
            self.print_board()
            if self.check_winner(self.computer_symbol):
                print("Sorry, you lose!")
                return
            if ' ' not in self.board:
                print("It's a draw!")
                return


firstGame = TicTacToeGame()
firstGame.play_game()
```

```python
#TASK 6

import heapq

my_graph = {
    'Start': {'Kitchen': 10, 'Garden': 20, 'LivingRoom': 10},
    'Kitchen': {'Bathroom': 20, 'Garden': 10},
    'Garden': {'Bathroom': 30, 'Kitchen': 8, 'Start': 20},
    'LivingRoom': {'Bathroom': 15, 'Kitchen': 10, 'Start': 10, 'Garden': 8},
    'Bathroom': {}
}

start_point = 'Start'
end_point = 'Bathroom'

distances = {node: float('inf') for node in my_graph}
distances[start_point] = 0

previous_nodes = {node: None for node in my_graph}

priority_queue = [(0, start_point)]

while priority_queue:
    current_dist, current_node = heapq.heappop(priority_queue)

    if current_node == end_point:
        break

    for neighbor in my_graph[current_node]:
        new_dist = current_dist + my_graph[current_node][neighbor]

        if new_dist < distances[neighbor]:
            distances[neighbor] = new_dist
            previous_nodes[neighbor] = current_node
            heapq.heappush(priority_queue, (new_dist, neighbor))

chota_path = []
current_node = end_point

while current_node is not None:
    chota_path.append(current_node)
    current_node = previous_nodes[current_node]

chota_path.reverse()
print("The shortest path is:", chota_path)
print("The total distance is:", distances[end_point])
```
