The Fibonacci sequence is a sequence of numbers where the number is the sum of two
previous numbers. Write the Fibonacci sequence through recursion of 10.

```python
def Fibonacci(n1):

    if n1 < 0 :
        print ("Invalid value")

    elif n1 == 0:
        return 0;

    elif n1 == 1 or n1 == 2:
        return 1

    else:
        return Fibonacci(n1 - 1) + Fibonacci(n1 - 2)


for n in range (1,11):
    print(Fibonacci(n))
```

You are in a hurry and the traffic officer stops you. Write a function to return one of 3 possible results: "No fine", "Less Fine", or "Car seize". If your speed is 70 or less, the result is "No fine". If speed is between 71 and 80 inclusive, the result is "Less Fine". If speed is 81 or more, the result is "Car seize". Unless it is your anniversary (encoded as a Boolean value in the parameters of the function) -- on your anniversary, your speed can be 10 higher in all cases

```python
def fineFunc(speed, isAnniversary):
  if isAnniversary:
    if speed <= 80:
      print("No fine")
    elif speed >= 81 and speed <= 90:
      print("Less fine")
    else:
      print("Car Seize")
  else:
    if speed <= 70:
      print("No fine")
    elif speed >=71 and speed <= 80:
      print("Less fine")
    else:
      print("Car seize")

speed = int(input("Please tell me your speed? "))
isAnniv = input("Is it driver's anniversary today? (True/False): ")
flag = isAnniv.lower() == "true"
if flag:
  fineFunc(speed,flag)
else:
  fineFunc(speed,flag)
```

Determine whether your data contains any duplicate element in a given array of integers.
Return true if any value appears twice in the array and return false if every element is
distinct. Also, remove the duplicates.

```python
def duplicate_present(arr):
    element_count = {}
    contain_duplicates = False
    distinct_elements = []

    for num in arr:
        if num in element_count:
            element_count[num] += 1
            contain_duplicates = True
        else:
            element_count[num] = 1
            distinct_elements.append(num)

    return contain_duplicates, distinct_elements

arr = [1, 2, 3, 4, 5, 6, 7, 7]
contain_duplicates, distinct_elements = duplicate_present(arr)

print(f"Duplicates Present? {contain_duplicates}")
print(f"Data with distinct elements: {distinct_elements}")
```

 A user enters any string containing letters and digits both. Calculate the number of letters and digits entered by a user.

```python
userInp = input("Enter mix of letters and words: ")

digits = sum(i.isdigit() for i in userInp)
letters = sum(i.isalpha() for i in userInp)

print(f"Number of digits is {digits}")
print(f"Number of letters is {letters}")
```
