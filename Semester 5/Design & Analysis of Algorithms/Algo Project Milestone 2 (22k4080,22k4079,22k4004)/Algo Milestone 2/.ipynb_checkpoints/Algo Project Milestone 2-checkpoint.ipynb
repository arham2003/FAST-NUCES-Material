# Q2 - Generating Input Files

```python
import random
import os

def generate_points(num_points):
    return [(random.randint(-1000, 1000), random.randint(-1000, 1000)) for _ in range(num_points)]

def generate_large_integers(size):
    num1 = ''.join(random.choices('0123456789', k=size))
    num2 = ''.join(random.choices('0123456789', k=size))
    return num1, num2

output_dir = "generated_inputs"
os.makedirs(output_dir, exist_ok=True)

for i in range(1, 11):
    num_points = random.randint(101, 500) 
    points = generate_points(num_points)
    file_path = os.path.join(output_dir, f"closest_pair_input_{i}.txt")
    with open(file_path, 'w') as f:
        for point in points:
            f.write(f"{point[0]} {point[1]}\n")

for i in range(1, 11):
    size = random.randint(1, 50)
    num1, num2 = generate_large_integers(size)
    file_path = os.path.join(output_dir, f"integer_multiplication_input_{i}.txt")
    with open(file_path, 'w') as f:
        f.write(f"{num1}\n{num2}\n")

print(f"Generated inputs saved in '{output_dir}' directory.")
```

## Q3 (Applying closest pair and Integer Multiplication on The generated Input Files)

```python
import os
import math

def closest_pair(points):
    def dist(p1, p2):
        return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
    
    def closest_split_pair(px, py, delta, best_pair):
        mid_x = px[len(px) // 2][0]
        sy = [p for p in py if mid_x - delta <= p[0] <= mid_x + delta]
        best = delta
        len_sy = len(sy)
        for i in range(len_sy):
            for j in range(i + 1, min(i + 8, len_sy)):
                p, q = sy[i], sy[j]
                d = dist(p, q)
                if d < best:
                    best = d
                    best_pair = (p, q)
        return best_pair, best
    
    def closest_pair_recursive(px, py):
        if len(px) <= 3:
            min_dist = float('inf')
            best_pair = None
            for i in range(len(px)):
                for j in range(i + 1, len(px)):
                    d = dist(px[i], px[j])
                    if d < min_dist:
                        min_dist = d
                        best_pair = (px[i], px[j])
            return best_pair, min_dist
        
        mid = len(px) // 2
        Qx, Rx = px[:mid], px[mid:]
        midpoint = px[mid][0]
        Qy, Ry = [], []
        for point in py:
            if point[0] <= midpoint:
                Qy.append(point)
            else:
                Ry.append(point)
        
        (p1, q1), dist1 = closest_pair_recursive(Qx, Qy)
        (p2, q2), dist2 = closest_pair_recursive(Rx, Ry)
        
        if dist1 <= dist2:
            d = dist1
            best_pair = (p1, q1)
        else:
            d = dist2
            best_pair = (p2, q2)
        
        (p3, q3), dist3 = closest_split_pair(px, py, d, best_pair)
        
        if d <= dist3:
            return best_pair, d
        else:
            return (p3, q3), dist3
    
    px = sorted(points, key=lambda p: p[0])
    py = sorted(points, key=lambda p: p[1])
    return closest_pair_recursive(px, py)


def karatsuba(x, y): ## Karatsuba's Algorithm for Integer Multiplication
    if x < 10 or y < 10:  # Base case
        return x * y
    
    n = max(len(str(x)), len(str(y)))
    half = n // 2
    
    high_x, low_x = divmod(x, 10**half)
    high_y, low_y = divmod(y, 10**half)
    
    z0 = karatsuba(low_x, low_y)
    z1 = karatsuba(low_x + high_x, low_y + high_y)
    z2 = karatsuba(high_x, high_y)
    
    return (z2 * 10**(2 * half)) + ((z1 - z2 - z0) * 10**half) + z0


input_dir = "generated_inputs"
output_dir = "generated_outputs"
os.makedirs(output_dir, exist_ok=True)

for i in range(1, 11):
    input_file = os.path.join(input_dir, f"closest_pair_input_{i}.txt")
    output_file = os.path.join(output_dir, f"closest_pair_output_{i}.txt")
    
    points = []
    with open(input_file, 'r') as f:
        for line in f:
            x, y = map(int, line.split())
            points.append((x, y))
    
    closest_points, min_distance = closest_pair(points)
    with open(output_file, 'w') as f:
        f.write(f"Closest Points: {closest_points}\n")
        f.write(f"Minimum Distance: {min_distance:.6f}\n")


for i in range(1, 11):
    input_file = os.path.join(input_dir, f"integer_multiplication_input_{i}.txt")
    output_file = os.path.join(output_dir, f"integer_multiplication_output_{i}.txt")
    
    with open(input_file, 'r') as f:
        num1 = int(f.readline().strip())
        num2 = int(f.readline().strip())
    
    product = karatsuba(num1, num2)
    with open(output_file, 'w') as f:
        f.write(f"Product: {product}\n")

print(f"Output files saved in '{output_dir}' directory.")
```

# Q4 GUI For Visualizing the Algorithms

```python
import tkinter as tk
from tkinter import filedialog, messagebox, Canvas
import math
import time

def closest_pair_window():
    def load_file():
        file_path = filedialog.askopenfilename(title="Select Input File", filetypes=[("Text files", "*.txt")])
        if not file_path:
            return
        with open(file_path, "r") as f:
            points = [tuple(map(int, line.strip().split())) for line in f.readlines()]
        visualize_closest_pair(points)

    def visualize_closest_pair(points):
        canvas.delete("all")
        canvas.create_text(400, 20, text="Visualizing Closest Pair Algorithm", font=("Arial", 14))

        sorted_points = sorted(points, key=lambda p: p[0])
        for i, (x, y) in enumerate(sorted_points):
            canvas.create_oval(x + 10, y + 10, x + 15, y + 15, fill="blue")
            canvas.create_text(x + 15, y, text=f"{i}")
        canvas.update()

        time.sleep(2)

        pair, dist = closest_pair(sorted_points)
        canvas.create_line(pair[0][0] + 10, pair[0][1] + 10, pair[1][0] + 10, pair[1][1] + 10, fill="red", width=2)
        canvas.create_text(400, 450, text=f"Closest Pair: {pair} with Distance: {dist:.2f}", font=("Arial", 12))
        canvas.update()

    window = tk.Toplevel(root)
    window.title("Closest Pair Visualization")
    window.geometry("1000x800")
    window.configure(bg="dark grey") 

    load_button = tk.Button(
        window, text="Load Input File", command=load_file, font=("Arial", 12), bg="white", fg="black"
    )
    load_button.pack()

    canvas = tk.Canvas(window, width=800, height=600, bg="white")
    canvas.pack()


def integer_multiplication_window():
    def load_file():
        file_path = filedialog.askopenfilename(title="Select Input File", filetypes=[("Text files", "*.txt")])
        if not file_path:
            return
        with open(file_path, "r") as f:
            num1 = int(f.readline().strip())
            num2 = int(f.readline().strip())
        visualize_karatsuba(num1, num2)

    def visualize_karatsuba(num1, num2):
        canvas.delete("all")
        canvas.create_text(400, 20, text=f"Visualizing Karatsuba Algorithm\n{num1} x {num2}", font=("Arial", 14))
        canvas.update()

        result = karatsuba(num1, num2)
        time.sleep(2)

        canvas.create_text(400, 450, text=f"Result: {result}", font=("Arial", 16), fill="green")
        canvas.update()

    window = tk.Toplevel(root)
    window.title("Integer Multiplication Visualization")
    window.geometry("1000x800")
    window.configure(bg="dark grey") 

    load_button = tk.Button(
        window, text="Load Input File", command=load_file, font=("Arial", 12), bg="white", fg="black"
    )
    load_button.pack()

    canvas = tk.Canvas(window, width=800, height=600, bg="white")
    canvas.pack()


root = tk.Tk()
root.title("Algorithm Visualizer")
root.geometry("400x200")
root.configure(bg="dark green")  

title_label = tk.Label(root, text="Algorithm Visualizer", font=("Arial", 18), bg="dark green", fg="white")
title_label.pack(pady=10)

closest_pair_button = tk.Button(
    root, text="Closest Pair", command=closest_pair_window, font=("Arial", 14), width=15, bg="white", fg="black"
)
closest_pair_button.pack(pady=10)

integer_multiplication_button = tk.Button(
    root, text="Integer Multiplication", command=integer_multiplication_window, font=("Arial", 14), width=20, bg="white", fg="black"
)
integer_multiplication_button.pack(pady=10)

root.mainloop()
```
