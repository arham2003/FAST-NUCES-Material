```python
#task 5 (labeled as task 6)

legend_labels = ["Hyrule","Kakariko","Gerudo Valley"]
hyrule = [63, 65, 68, 70, 72, 72, 73]
kakariko = [52, 36, 45, 68, 39, 17, 65]
gerudu = [10, 12, 21, 42, 56, 32, 78]
Days = range(1,8)

plt.plot(Days,hyrule,label=legend_labels[0])
plt.plot(Days,kakariko,label=legend_labels[1])
plt.plot(Days,gerudu,label=legend_labels[2])
plt.legend(loc=8)
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.show()
```
