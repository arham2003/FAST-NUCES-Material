```python
#TASK 1

initial_state = [' ' for _ in range(9)]

def print_board(state):
    print(f"{state[0]} | {state[1]} | {state[2]}\n----------\n{state[3]} | {state[4]} | {state[5]}\n----------\n{state[6]} | {state[7]} | {state[8]}\n\n")

# Function to check if the game is over
def is_game_over(state):
    win_conditions = [(0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6)]
    for condition in win_conditions:
        if state[condition[0]] == state[condition[1]] == state[condition[2]] != ' ':
            return True, state[condition[0]]
    if ' ' not in state:
        return True, 'Tie'
    return False, None

# Function to evaluate the score of the board state
def evaluate(state):
    if is_game_over(state)[1] == 'X':
        return -10
    elif is_game_over(state)[1] == 'O':
        return 10
    return 0


def generate_possible_moves(state, player):
    moves = []
    for i in range(len(state)):
        if state[i] == ' ':
            new_state = state[:]
            new_state[i] = player
            moves.append(new_state)
    return moves

# min max function
def minimax(state, depth, is_maximizing):
    if is_game_over(state)[0]:
        return evaluate(state)

    if is_maximizing:
        best_score = float('-inf')
        for move in generate_possible_moves(state, 'O'):
            score = minimax(move, depth + 1, False)
            best_score = max(score, best_score)
        return best_score
    else:
        best_score = float('inf')
        for move in generate_possible_moves(state, 'X'):
            score = minimax(move, depth + 1, True)
            best_score = min(score, best_score)
        return best_score

def find_best_move(state, player):
    best_move = None
    if player == 'O':
        best_score = float('-inf')
        for move in generate_possible_moves(state, player):
            score = minimax(move, 0, False)
            if score > best_score:
                best_score = score
                best_move = move
    else:
        best_score = float('inf')
        for move in generate_possible_moves(state, player):
            score = minimax(move, 0, True)
            if score < best_score:
                best_score = score
                best_move = move
    return best_move

current_state = initial_state
current_player = 'X' # or 'O'
while not is_game_over(current_state)[0]:
    best_move = find_best_move(current_state, current_player)
    current_state = best_move
    print_board(current_state)
    current_player = 'O' if current_player == 'X' else 'X'
```

```python
#TASK 2

# Alpha-Beta Pruning Algorithm

def alpha_beta_pruning(node, depth, alpha, beta, maximizing_player):
    if depth == 0 or isinstance(node, int):
        return node

    if maximizing_player:
        max_eval = float('-inf')
        for child in node:
            eval = alpha_beta_pruning(child, depth - 1, alpha, beta, False)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break
        return max_eval
    else:
        min_eval = float('inf')
        for child in node:
            eval = alpha_beta_pruning(child, depth - 1, alpha, beta, True)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break
        return min_eval

# Example tree structure based on the provided image
tree = [
    [
        [2, 4], [6, 8]
    ],
    [
        [1, 2], [10, 12]
    ]
]

# Call the alpha-beta pruning algorithm
# Assuming the root is a maximizing node and the depth is 3
optimal_value = alpha_beta_pruning(tree, 3, float('-inf'), float('inf'), True)
print(f"The optimal value is: {optimal_value}")
```

```python
#task 3

def is_safe_to_move(board, row, col):
    # Check this row on left side
    for i in range(col):
        if board[row][i] == 1:
            return False

    # Check upper diagonal on left side
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    # Check lower diagonal on left side
    for i, j in zip(range(row, len(board), 1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    return True

def solve_n_queens_util(board, col):
    # base case: If all queens are placed
    if col >= len(board):
        return True

    # Consider this column and try placing this queen in all rows one by one
    for i in range(len(board)):
        if is_safe_to_move(board, i, col):
            # Place this queen in board[i][col]
            board[i][col] = 1

            # recur to place rest of the queens
            if solve_n_queens_util(board, col + 1):
                return True

            # If placing queen in board[i][col] doesn't lead to a solution,
            # then remove queen from board[i][col]
            board[i][col] = 0

    # if the queen can not be placed in any row in this column col then return false
    return False

def solve_n_queens(n):
    board = [[0 for _ in range(n)] for _ in range(n)]

    if not solve_n_queens_util(board, 0):
        print("Solution does not exist")
        return False

    # Print the solution
    for i in range(n):
        for j in range(n):
            print("Q" if board[i][j] else ".", end=" ")
        print()
    return True


n = 4  # value for different sizes (NO. OF QUEENS = BOARD SIZE)
solve_n_queens(n)
```

```python
#TASK 4

from itertools import permutations

def sol_cryptarithmetic(addends, result):

    unique_chars = set(''.join(addends) + result)
    if len(unique_chars) > 10:
        return "Too many letters for a valid cryptarithmetic puzzle."

    # Generate all possible mappings of characters to digits
    for perm in permutations('0123456789', len(unique_chars)):
        solution = dict(zip(unique_chars, perm))

        # Skip solutions that assign 0 to a leading character
        if any(solution[word[0]] == '0' for word in addends + [result]):
            continue

        # Check if the current mapping is a solution
        if sum(int("".join(solution[c] for c in word)) for word in addends) == int("".join(solution[c] for c in result)):
            return solution

    return "No solution found."


addends = ["BASE", "BALL"]
result = "GAMES"
solution = sol_cryptarithmetic(addends, result)

if isinstance(solution, str):
    print(solution)
else:
    print("Solution found:")
    for k, v in solution.items():
        print(f"{k} = {v}")
```
