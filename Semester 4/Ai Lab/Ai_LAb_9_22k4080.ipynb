## Task 1 - **LAB 9**

```python
sample_space = ['BBBB', 'BBBG', 'BBGB', 'BBGG', 'BGBB', 'BGBG', 'BGGB', 'BGGG',
                'GBBB', 'GBBG', 'GBGB', 'GBGG', 'GGBB', 'GGBG', 'GGGB', 'GGGG']

outcomes_two_boys = [outcome for outcome in sample_space if outcome.count('B') == 2]

probability_of_two_boys = len(outcomes_two_boys) / len(sample_space)

print("Sample Space (Total Outcomes):")
print(sample_space)
print("\nOutcomes with Exactly Two Boys:")
print(outcomes_two_boys)
print("\nProbability of Having Exactly Two Boys:")
print(f"{probability_of_two_boys:.3f}")
```

## Task 2 - **LAB 9**

```python
probabilities = {
    '1': 1/9,
    '2': 2/9, # even numbers have twice the chance
    '3': 1/9,
    '4': 2/9,
    '5': 1/9,
    '6': 2/9
}

sample_space = ['1', '2', '3', '4', '5', '6'] # sample space with the biased probabilities

P_E = probabilities['1'] + probabilities['2'] + probabilities['3']

print("Sample Space (Total Outcomes):")
print(sample_space)

print("\nProbabilities for Each Outcome:")
for outcome, probability in probabilities.items():
    print(f"Outcome {outcome}: {probability}")

print("\nProbability of Event E (Rolling a Number Less Than 4):")
print(f"{P_E:.2f}")
```

```python
from hmmlearn import hmm

model = hmm.CategoricalHMM(n_components=2)

observations = [[0], [1]]  # 0: no cough, 1: cough

model.fit(observations)

p_healthy = model.startprob_[0]

print("Model Parameters:")
print("---------------------")
print("Initial state probabilities (healthy, sick):\n\n", model.startprob_)
print("Transition probabilities (healthy -> healthy, sick -> sick):\n\n", model.transmat_)
print("Observation probabilities (healthy: no cough, sick: cough):\n\n", model.emissionprob_)
print("\nProbability that the person is healthy:", p_healthy)
```

```python
prob_plan = {'P1': 0.3, 'P2': 0.2, 'P3': 0.5}

prob_defect_given_plan = {'P1': 0.01, 'P2': 0.03, 'P3': 0.02}

prob_defect = (prob_plan['P1'] * prob_defect_given_plan['P1'] +
               prob_plan['P2'] * prob_defect_given_plan['P2'] +
               prob_plan['P3'] * prob_defect_given_plan['P3'])

prob_plan_given_defect = {
    plan: (prob_plan[plan] * prob_defect_given_plan[plan]) / prob_defect
    for plan in prob_plan
}

most_likely_plan = max(prob_plan_given_defect, key=prob_plan_given_defect.get)# plan with the highest probability given a defect

print("Probability of each plan given a defect:")
for plan, probability in prob_plan_given_defect.items():
    print(f"{plan}: {probability:.4f}")

print("\nMost likely plan responsible for a defective product:", most_likely_plan)
```
