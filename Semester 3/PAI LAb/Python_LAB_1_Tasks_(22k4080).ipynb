```python
#Q3 22k-4080
import random

userNum = int(input('Input any number: \n'))

if userNum <= 0:
  print("Please enter a positive number.")

unique_numbers = []

while len(unique_numbers) < userNum:
    random_num = random.randint(1,userNum)

    if random_num not in unique_numbers :
        unique_numbers.append(random_num)

print("Random number list: \n")
print(unique_numbers)
```

```python
#Q4 22k-4080

import random


def play_number_guessing_game():
    while True:

        target_number = random.randint(1, 100)
        attempts = 3

        print("Welcome to the Number Guessing Game!")
        print("I'm thinking of a number between 1 and 100.")

        while attempts > 0:

            user_guess = int(input(f"You have {attempts} attempts left. Enter your guess: "))


            if user_guess == target_number:
                print("Congratulations! You guessed the correct number.")
                break
            elif user_guess < target_number:
                print("Too low. Try again.")
            else:
                print("Too high. Try again.")

            attempts -= 1

        if attempts == 0:
            print(f"Sorry, you've run out of attempts. The correct number was {target_number}.")


        play_again = input("Do you want to play again? (y/n): ").strip()

        if play_again != 'y':
            print("Thanks for playing! Goodbye.")
            break

if __name__ == "__main__":
    play_number_guessing_game()
```

```python
#Q5 22k-4080

print ("--Welcome to BMI calculator!--\n")
weight = float(input("Input Weight: "))
height = float(input("Input height (m): "))
BMI = float(0)

BMI = float(weight /(height)**2)
print(f"BMI: {round(BMI, 2)}")
if (BMI < 18.5) :
    print("Underweight!")
elif (BMI >= 18.5 and BMI <= 25):
    print("Normal weight")
elif (BMI > 25 and BMI <= 30) :
    print("obese")
else:
    print("Clinically obese")

```

```python
#Q9 22k-4080

def main():
  print("--Leap Year Calculator--")
  userYear = int(input("Please enter the year you want to check for leap year "))
  leapYear(userYear)

def leapYear(year):
  if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("The year", year, "is a leap year")
        else:
            print("The year", year, "is not a leap year")
    else:
        print("The year", year, "is a leap year")
  else:
    print("The year", year, "is not a leap year")


if __name__ == "__main__":
    main()
```
